 Data Visualization I

    1. Use the inbuilt dataset 'titanic'. The dataset contains 891 rows and contains information 
    about the passengers who boarded the unfortunate Titanic ship. Use the Seaborn library to 
    see if we can find any patterns in the data. 
    2. Write a code to check how the price of the ticket (column name: 'fare') for each passenger 
    is distributed by plotting a histogram.


```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the 'titanic' dataset from Seaborn
df = sns.load_dataset('titanic')

# Preview the first 5 rows of the dataset
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['age'] = df['age'].fillna(np.mean(df['age']))
df['embarked'].fillna(df['embarked'].mode()[0],inplace=True)
df['deck'].fillna(df['deck'].mode()[0],inplace=True)
df['embark_town'].fillna(df['embark_town'].mode()[0],inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head()
```

# Data Visualization

```python

sns.histplot(x='fare', data=df)
```

This histogram shows how the price of the ticket for each passenger is distributed

```python
plt.figure(figsize=(5,5))
sns.histplot(x='age', data=df)
```

This histogram shows how the price of the ticket for each passenger is distributed

```python
df.hist(column='fare')
```

```python
sns.displot(df['age'],bins=10,kde=True)
```

```python
sns.displot(x=df['age'],bins=10)
```

```python
sns.jointplot(x='age',y='fare',data = df)
```

```python
sns.jointplot(x=df['age'],y=df['fare'],kind='hex')
```

```python
sns.rugplot(df['fare'])
```

```python
# Bar plot
sns.barplot(x='sex',y='age',data=df)
```

```python
# Count plot
sns.countplot(x='sex',data=df)
```

```python
# Box plot
sns.boxplot(x='sex',y='age',data=df)
```

```python
sns.boxplot(x='sex',y='age',data=df,hue="survived")
```

```python
#Violin plot
sns.violinplot(x='sex',y='age',data=df)
```

```python
#Violin plot
sns.violinplot(x='sex',y='age',data=df,hue='survived')
```

```python
sns.stripplot(x="sex",y="age",data=df,jitter =True)
```

```python
#swarm plot
sns.swarmplot(x='sex',y='age',data=df)
```
