```python
# Refer handout for explanation
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
# Load the 'titanic' dataset from Seaborn
df = sns.load_dataset('titanic')

# Preview the first 5 rows of the dataset
df.head()
```

```python
# DIst plot
sns.displot(df['age'],bins=10,kde=True)

```

```python
sns.displot(x=df['age'],bins=10)



```

```python
#joint plot
sns.jointplot(x=df['age'],y=df['fare'],kind='scatter')
```

```python
#joint plot
sns.jointplot(x=df['age'],y=df['fare'],kind='hex')
```

```python
# Rug plot
sns.rugplot(df['fare'])
```

```python
# Bar plot
sns.barplot(x='sex',y='age',data=df)
```

```python
import numpy as np
sns.barplot(x='sex',y='age',data=df,estimator = np.std)
```

```python
# Count plot
sns.countplot(x='sex',data=df)
```

```python
# Box plot
sns.boxplot(x='sex',y='age',data=df)
```

```python
sns.boxplot(x='sex',y='age',data=df,hue="survived")
```

```python
#Violin plot
sns.violinplot(x='sex',y='age',data=df)
```

```python
sns.violinplot(x='sex',y='age',data=df,hue="survived")
```

```python
#Strip plot
sns.stripplot(x="sex",y="age",data=df,jitter =False)
```

```python
sns.stripplot(x="sex",y="age",data=df,jitter =True)
```

```python
sns.stripplot(x="sex",y="age",data=df,jitter =True,hue="survived") 
```

```python
#swarm plot
sns.swarmplot(x='sex',y='age',data=df)
```

```python
sns.swarmplot(x='sex',y='age',data=df,hue='survived')
```

```python
corr=df.corr()
sns.heatmap(corr)
```

```python
corr=df.corr()
sns.heatmap(corr,annot=True)
```

```python
#cluster map
sns.histplot(df['fare'],kde=False,bins=10)
```
